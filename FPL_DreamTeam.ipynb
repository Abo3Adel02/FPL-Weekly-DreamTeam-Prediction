### The Data

The Data We will be using is openly available on Github and anyone can access it [here](https://github.com/vaastav/Fantasy-Premier-League/tree/master/data/2021-22). Entire of the data is players data in the 2021/2022 Season

#### FPL_Data.csv

-----

`FPL_Data.csv` contains players data in every GameWeek for the 2021/2022 Season. Players Data inlcuding his name, position, team, opponent team, his statistics in a particular GameWeek, information about the fixture, and his value. The data's latest update was on Aug. 02, 2022.

**NOTE:** The Featrue Description Below is for the New and Updated Dataset that will be used for analysis,
visualization, & modeling.

Feature | Definition
---- | -----------
Player_Name | Player's name
Player_Position | Player's role on the pitch
Team_Level | Player's current team level he is playing for
Player_Value | Player's current price in-game (in millions)
GameWeek | Current week number 
Fixture_Difficulty_Rate | Player's current fixture Difficulty
Was_Home | Is the current game being played in his team stadium?
Player_Form | Avarage points scored by the player last 4 games
Minutes_Played | Number of minutes the player played the game
Goals_Scored | Number of goals scored by the player
Goals_Conceded | Number of goals the opposition team scored
Assists | Number of direct passes that led to a goal
Bonus | Number of additional points the player gets according to his performance on the game
BPS | Bonus Point System (System made by the game itself to measure how well the player played
Influence | Number indicating how much of an impact the played had in the game
Creativity | Number indicating the creative passes or dribbling the player made
Threat | Number indicating how much of a threat (Key passes/ Key Dribbles/Goals) did the player do to the opponent team
ICT_Index | Statistical Feature made by the game itself combining (Influence-Creativity-Threat) All together determining the form of the player
Saves | Number of shots on target saved by the goalkeeper
Penalties_Saved | Number of saved penalties the goalkeeper did in the game
Penalties_Missed | Number of penalties missed by the player in the game
Clean_Sheet | Did the team got a clean sheet (opponent team scored zero goals)?
Total_Points | Number of total points the player got in the game
Team_Strength_Overall_Home | Player's team overall strength at home
Team_Strength_Overall_Away | Player's team overall strength away
Next_Fixture_Difficulty_Rate | Player's upcoming fixture difficulty
In_DreamTeam | Will the player be selected in the upcoming GameWeek DreamTeam?

### 5.2.1 Create a Pipeline to Prevent Data Leakage

**OneHotEncoding Categorical Data**

To Encode the nominal categorical Data

```python
import category_encoders as OHE

Encoder = OHE.OneHotEncoder(cols=['Player_Position'])
```

**RobustScaler for Outliers**

To Scale the features espacially outliers with an insensitive to outliers scaler

```python
from sklearn.preprocessing import RobustScaler

Rscaler = RobustScaler()
```

**StandardScaler**

To Standardize the features after dealing with outliers, to make computational cost less

```python
from sklearn.preprocessing import StandardScaler

Sscaler = StandardScaler()
```

**LogisticRegression Model**

Declaring the model to perform GridSearchCV on it and its Hyperparameters

```python
from sklearn.linear_model import LogisticRegression

LGR = LogisticRegression(solver='saga', max_iter=1000)
```

**Declaring the Pipeline**

After creating the required encoding, scaling, and modeling operations, now create the list of operations that is meant to happen sequential

```python
operations = [('ROS', ROS), ('Encoder', Encoder), ('Rscaler', Rscaler), ('Sscaler', Sscaler), ('LGR', LGR)]
```

```python
from imblearn.pipeline import Pipeline

Pipe = Pipeline(operations)
```

**MODEL SELECTION: Finding the best Hyperparameters Using GridSearchCV**

```python
Param_Grid = {'LGR__penalty': ['elasticnet', 'l1', 'l2', 'none'], 'LGR__C': [0.0000001, 0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1], 
             'LGR__l1_ratio': [0, 0.1, 0.3, 0.5, 0.7, 0.9, 0.95, 0.99, 1, 10]}
```

```python
from sklearn.model_selection import GridSearchCV

Grid_Model = GridSearchCV(Pipe, param_grid=Param_Grid, cv=5, scoring='precision', n_jobs=-1)

Grid_Model.fit(X, y)
```
